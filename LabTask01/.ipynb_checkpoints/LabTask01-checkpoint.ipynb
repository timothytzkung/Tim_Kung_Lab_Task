### Part 5: Arithmetic and Data Alignment ###

1. Create a new column that converts fares from USD to Euros.

2. Subtract the average age from every age value.

3. Subtract a custom array from two selected columns in the dataset.

### 5.1 Solution
#### Q: Create a new column that converts fares from USD to Euros.

```python
# Assuming Current Fares are in USD; where 1USD => 0.86 Euro
df["Fare (Euros)"] = df["Fare"] * 0.86
df
```

### 5.2 Solution
#### Q: Subtract the average age from every age value.

```python
# To avoid constant subtraction of age
_df = df.copy()
avg_age = _df.loc[:, "Age"].mean()
print(f"The average age is: {avg_age} years old")

_df["Age"] = _df["Age"] - avg_age

# Variations
# _df => Returns raw data frame with negative age included

# Returns cleaned data frame where age is non-negative and non-null
_df[ (_df["Age"] > 0) & (~_df["Age"].isnull()) ]
```

### 5.3 Solution
#### Q: Subtract a custom array from two selected columns in the dataset.
##### Note: Subtracts average Euro fare from Euro & USD Fare

```python
_df = df.copy()

# Gets average price
avg_price = _df["Fare (Euros)"].mean()
print("Average Euro Fare is: ", avg_price)

# Creates array with avg euros
custom_array = np.full(2,avg_price)

# Apply Subtraction
_df[["Fare", "Fare (Euros)"]] -= custom_array
# if want non-clean, can just call "_df"

# Cleaning for negative prices
_df[ (_df["Fare"] >= 0) & (_df["Fare (Euros)"] >= 0) ]
```

### Part 6: Applying Functions ###

1. Create a new column with the number of characters in each name.

2. Calculate the difference between fare and age for each row.

3. Round all numerical values in the dataset to one decimal place.



### 6.1 Solution
#### Q: Create a new column with the number of characters in each name.

```python
_df = df.copy()

# Apply cleaning to only include alphabetical letters/characters
_df["Number of Characters (Name)"] = _df["Name"].apply(lambda x: len("".join(c for c in x if c.isalpha())))
_df
```

### 6.2 Solution
#### Q: Calculate the difference between fare and age for each row.

```python
_df = df.copy()

# Calculates diff in age if Age not null & fares not negative (valid fares)
_df["Difference (Fare - Age)"] = df["Fare"] - df["Age"]
_df[ (~_df["Age"].isnull()) & (_df["Fare"] >= 0) & (_df["Fare (Euros)"] >= 0)]
```

### 6.3 Solution
#### Q: Round all numerical values in the dataset to one decimal place.

```python
_df = df.copy()

_df.round(1)
```

### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



### 7.1 Solution
#### Q: Convert categorical values in the gender column to numeric codes.

```python

# Replace categorical => numerical for sex
_df = df.copy()

# Updates column with new values (numeric code)
_df["Sex"] = _df["Sex"].replace({"female": 0, "male": 1})
_df
```

### 7.2 Solution
#### Q: Create a new column that categorizes age into groups (e.g., child, adult, senior).

```python
# Assume the following mapping for child, adult, senior
# Child <= 18, Adult 18 ~ 65, Senior 65+

_df = df.copy()

# Format of naming
format = lambda x: "Child" if x < 18 else ("Senior" if x > 65 else "Adult")

# Remove null vals in age
_df = _df[ ~df["Age"].isnull() ] 

# Apply format
_df["Age Category"] = _df["Age"].map(format)

# Display top 30 results (1 senior at passengerId 34)
_df.head(30)
```
